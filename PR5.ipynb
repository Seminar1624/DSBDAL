
**Data Visualization II
    1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names: 'sex' and 'age')**

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
# Create the boxplot

sns.boxplot(x='sex', y='age', hue='survived', data=titanic, palette='Set2')

plt.title('Age Distribution by Gender and Survival Status (1-Survived)', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Age', fontsize=12)

# Show the plot
plt.show()
```

**Analysis : The survival rate of female is more in average age from 20 to 40
the Unsurvival rate of male is more where the age between 20 to 40**

```python
sns.barplot(x='sex', y='age', hue='survived', data=titanic, palette='Set2')

# Add title and labels
plt.title('Average Age by Gender and Survival Status', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Age', fontsize=12)

# Show the plot
plt.show()
```

**Analysis : The survival rate of female is more in average age from 25 to 30
the Unsurvival rate of male is less where the age is above 30**
