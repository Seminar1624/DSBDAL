```python
# Check distribution of the target variable (e.g., 'Gender')
target_distribution = data['Gender'].value_counts()
print(target_distribution)
```

```python
if target_distribution.min()/target_distribution.max() > 0.5:
    print("\nThe dataset is balanced.")
else:
     print("\nThe dataset is imbalanced.")
    
```

```python
#describe each class of Species
def basic_statistics(df):
    print("\nIris-setosa")
    print(df[df.Species=='Iris-setosa'].describe().transpose())
    print("\nIris-versicolor")
    print(df[df.Species=='Iris-versicolor'].describe().transpose())
    print("\nIris-virginica")
    print(df[df.Species=='Iris-virginica'].describe().transpose())
basic_statistics(iris_dataset)  
```
