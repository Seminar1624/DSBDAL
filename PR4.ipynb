```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv('titanic.csv')
data.head()


```

```python
plt.bar(data['Age'],data['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("FAre per Age")
plt.show()
```

```python
#plotting bar graph using x and y axix x=Survivied,y=Age
plt.bar(data["Survived"], data["Age"])
plt.xlabel("Survived")
plt.ylabel("Age")
plt.title("Age vs Survival")
```

```python
#Create a count plot to compare the distribution of age between survivors and non-survivors

sns.catplot(data=data, x='Survived', hue='Age', kind='count', height=5, aspect=1.5)
plt.title('Survival Distribution by Age')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

```python
#strip plot using pclass and age
sns.catplot(data=data, x='Pclass', y='Age', kind='strip', jitter=True, height=5, aspect=1.5)
plt.title('Individual Ages by Class (Strip Plot)')
plt.show()
```

```python
#Count of passengers by class
sns.catplot(data=data, x='Pclass', kind='count', height=5, aspect=1.5)
plt.title('Count of Passengers by Class')
plt.show()
```

```python
# Filter passengers who paid more than $5
high_fare_passengers = data[data['Fare'] > 5]

# Create a count plot to show the survival status of passengers who paid more than $5
sns.countplot(data=high_fare_passengers, x='Survived', palette={'0': 'red', '1': 'green'})
plt.title('Survival Status of Passengers Who Paid More Than $5')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count of Passengers')
plt.show()

# Print the survival statistics for passengers who paid more than $5
survival_rate = high_fare_passengers['Survived'].value_counts(normalize=True)
print(f"Survival Rate for Passengers Who Paid More Than $5:\n{survival_rate}")
```

**Analysis : Passengers who paid more than 5 were likely from various classes (including 2-class or some 3-class passengers), not necessarily just wealthy 1-class passengers. While it's true that 1-class passengers typically paid more and had a higher survival rate, there were also 3-class passengers who could have paid over** **5 for a ticket but had much lower survival rates due to their location on the ship, lack of access to lifeboats, or other factors.**

```python
#To analyze the number of females and males who survived and paid a fare 

# Filter out passengers who have paid a fare (Fare > 0) and survived
df_survived_paid_fare = data[(data['Fare'] > 0) & (data['Survived'] == 1)]

# Use seaborn to create a countplot for gender-based survival
#plt.figure(figsize=(8, 6))
sns.countplot(data=df_survived_paid_fare, x='Sex',palette={'female': 'red', 'male': 'green'})
plt.title('Survival by Gender (Paid Fare)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Analysis : The count of females who survived is higher than males, this suggests that women were more likely to survive among passengers who paid a fare.**

```python

# Filter out passengers who survived
df_survived = data[data['Survived'] == 1]

# Create a count plot to show survival based on the Embarked station
sns.countplot(data=df_survived, x='Embarked', palette='Set2')
plt.title('Survival by Embarkation Station')
plt.xlabel('Embarkation Station')
plt.ylabel('Count of Survived Passengers')
plt.show()

# Calculate the survival count based on Embarked
survival_by_embarked = df_survived['Embarked'].value_counts()
print("Survival Count by Embarkation Station:")
print(survival_by_embarked)
```

**Analysis : Embarkation station S has significantly more survivors compared to the other stations, it could indicate that the passengers from that station had a higher survival rate. This might be due to the location of lifeboats, the class distribution at each station, or even the type of passengers who boarded from that station.**

```python
#The scatter plot of bill lengths and bill by penguin species.
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=penguins, hue="species")
plt.title("Bill Length vs Bill Depth", size=20, color="red")
```

```python
#Histogram for flipper length

sns.histplot(x ="flipper_length_mm", data=penguins)
plt.title("Flipper Length", size=20, color="red")

```

**Analysis : 70 to 80 penguines have flipper length between 190-195(heighest count)**

```python
#histogram using kde value true

sns.histplot(x = "flipper_length_mm", data = penguins, kde = True)
plt.title("Flipper Length", size=20, color="red")

```

```python
#the bar plot showing the flipper lengths of penguin species.
sns.barplot(x ="species", y="flipper_length_mm", data=penguins)
plt.title("Flipper Length for 3 Penguin Species", size=20, color="red")
```

**Analysis : the species Gentoo has more Flipper length**

```python
#the hue parameter to see the flipper lengths of the species by sex
sns.barplot(x = "species", y = "flipper_length_mm", data =penguins, hue = "sex")
plt.title("Flipper Length for 3 Penguin Species by Sex", size=20, color="red")
```

**Analysis : the species Gentoo has more Flipper length of male**


```python
#The box plot 
#the distribution of flipper length by species.

sns.boxplot(x = "species", y = "flipper_length_mm", data =penguins)
plt.title("Flipper Length for 3 Penguin Species", size=20, color="red")
```

```python
#the violin plot as a box plot.
#the violin plot of flipper length.

sns.violinplot(x = "species", y = "flipper_length_mm", data = penguins)
plt.title("Flipper Length for 3 Penguin Species", size=20, color="red")

```

```python
#facet grid draw the histogram plot of the penguins flipper length according to the island and sex variables.

sns.FacetGrid(penguins, col="island", row="sex").map(sns.histplot, "flipper_length_mm")

```

```python
#headmap

# Select only numeric columns
numeric_penguins = penguins.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for the numeric columns and plot the heatmap
sns.heatmap(numeric_penguins.corr(), annot=True, cmap='coolwarm')

# Show the plot
plt.show()
```

Analysis : If flipper length and body mass have a high positive correlation (e.g., 0.8 or above), it suggests that penguins with longer flippers tend to be heavier. If bill length and bill depth have a weak or negative correlation, it means that a longer bill does not necessarily mean a deeper bill.
